```python
import numpy as np
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = nx.read_gml('./data/taxi-2014-NTA/12/NTACycle00Snap0.gml')
g = g.to_undirected()
```

```python
A = nx.adjacency_matrix(g).todense()
```

```python
eigvals = np.linalg.eigvalsh(A)
```

```python
print(min(eigvals),max(eigvals))
```
